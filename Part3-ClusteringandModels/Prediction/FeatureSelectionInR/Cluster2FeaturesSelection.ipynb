```r
data <- read.csv("~/Desktop/code/FavoritaGrocery/Part3-ClusteringandModels/Prediction/train_cluster2.csv")

set.seed(101)
indexes = sample(1:nrow(data), size=0.7*nrow(data))
train <- data[indexes, ]
test  <- data[-indexes, ]



lm.fit = lm(unit_sales~., data=train)
summary(lm.fit)
```

```r
p =predict(lm.fit, test)
library(forecast)
accuracy(p, test$unit_sales)
```

```r
library(leaps)
stepLM <- step(lm.fit, direction="both", steps = 500)
names(unlist(stepLM[[1]]))
```

```r
lm.fit3 = lm(unit_sales~onpromotion+transactions+dcoilwtico+day_0
             +day_1+day_2+day_3+day_4+month_1+month_2+month_3+month_4
             +month_7+month_8+month_9+month_10+Azuay+Bolivar
             +Chimborazo+Cotopaxi+El.Oro+Esmeraldas+Guayas+Imbabura
             +Loja+Los.Rios+Manabi+Pastaza+Pichincha+Santa.Elena
             +Santo.Domingo.de.los.Tsachilas+localholiday
             +regionalholiday+nationalother+nholidayspike+goodfriday
             +earthquakespike+earthquakedrop, data=train)
summary(lm.fit3)
```

```r
p3 =predict(lm.fit3, test)
library(forecast)
accuracy(p3, test$unit_sales)
```
