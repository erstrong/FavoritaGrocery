```r
data <- read.csv("~/Desktop/code/FavoritaGrocery/Part3-ClusteringandModels/Prediction/train_cluster5.csv")

set.seed(101)
indexes = sample(1:nrow(data), size=0.7*nrow(data))
train <- data[indexes, ]
test  <- data[-indexes, ]



lm.fit = lm(unit_sales~., data=train)
summary(lm.fit)
```

```r
lm.fit2 = lm(unit_sales~transactions+dcoilwtico+day_4+day_5+X1033+X1041
             +nationalother+nholidayspike+worldcupdrop, 
             data=train)
summary(lm.fit2)
```

```r
p2 =predict(lm.fit2, test)

library(forecast)
accuracy(p2, test$unit_sales)
```

```r
library(leaps)
regfit.bk = regsubsets(unit_sales~.,data=train,method="backward")
#summary(regfit.bk)
coef(regfit.bk,6)
```

```r
stepLM <- step(lm.fit, direction="both", steps = 1000)
names(unlist(stepLM[[1]]))
```

```r
lm.fit3 = lm(unit_sales~transactions+dcoilwtico+day_4+day_5++month_7
             +X1018+X1029+X1033+X1041+Pichincha+Santo.Domingo.de.los.Tsachilas
             +nationalother+nholidayspike+worldcupdrop, 
             data=train)
summary(lm.fit3)
```

```r
p3 =predict(lm.fit3, test)

accuracy(p3, test$unit_sales)
```
