```r
data <- read.csv("~/Desktop/code/FavoritaGrocery/Part3-ClusteringandModels/Prediction/train_cluster4.csv")

set.seed(101)
indexes = sample(1:nrow(data), size=0.7*nrow(data))
train <- data[indexes, ]
test  <- data[-indexes, ]



lm.fit = lm(unit_sales~., data=train)
summary(lm.fit)
```

```r
lm.fit2 = lm(unit_sales~onpromotion+transactions+dcoilwtico+day_4+day_5+month_1+month_2+month_3+month_4+month_5+month_6+month_7+month_8+month_9+month_10+month_11+X1025+X1028+X1032+X1036+X1038+X1039+X1044+X1048+X1054+X1056+X1078+X1086+X1092+Azuay+Bolivar+Cotopaxi+El.Oro+Guayas+Imbabura+Los.Rios+Manabi+Pastaza+Santa.Elena+Santo.Domingo.de.los.Tsachilas+nationalother+nholidayspike+earthquakespike, data=train)
summary(lm.fit2)
```

```r
p2 =predict(lm.fit2, test)

library(forecast)
accuracy(p2, test$unit_sales)
```
