```r
data <- read.csv("~/Desktop/code/FavoritaGrocery/Part3-ClusteringandModels/Prediction/train_cluster3.csv")

set.seed(101)
indexes = sample(1:nrow(data), size=0.7*nrow(data))
train <- data[indexes, ]
test  <- data[-indexes, ]



lm.fit = lm(unit_sales~., data=train)
summary(lm.fit)
```

```r
p =predict(lm.fit, test)
library(forecast)
accuracy(p, test$unit_sales)
```

```r
lm.fit2 = lm(unit_sales~.-day_6-month_5-month_12-X1045-Tungurahua-localholiday
             -worldcupdrop-earthquakespike-earthquakedrop-goodfriday, data=train)
summary(lm.fit2)
```

```r
p2 =predict(lm.fit2, test)

accuracy(p2, test$unit_sales)
```

```r
library(leaps)
stepLM <- step(lm.fit, direction="both", steps = 500)
names(unlist(stepLM[[1]]))
```
